## Marks and Channels

Marks and channels are the elements of the visual space that are used to represent our data. A mark is the visual analogue of a unit: it is a visual object that our visual perception picks out as an individual. A channel is the visual analogue of a variable: it is a way that a mark can vary.

**Some Common Marks**
- Points
- Lines
- Geometric shapes
- Curves

**Some Common Channels**
- Vertical position
- Horizontal position
- Length
- Width
- Angle
- Orientation
- Area
- Curvature
- Shape
- Hue
- Saturation
- Transparency
- Texture
- Connectedness
- Containment

Neither of these lists is exhaustive.

# Data Visualization

## Why visualize?

The plot below shows a famous set of datasets known as [Anscombe's quartet](https://en.wikipedia.org/wiki/Anscombe%27s_quartet). Anscombe's quartet contains four sets of pairs of points. The interesting thing about the quartet is that in each dataset, the following things are all true:

- The mean of $x$ is equal to 9
- The mean of $y$ is equal to 7.5
- The variance of $x$ is 11
- The variance of $y$ is 4.125
- The linear regression line with $y$ as a function of $x$ is $y = 3 + 5 x$
- The correlation coefficient between $y$ and $x$ 0.67

So if you were to perform some standard methods of analyzing and summarizing this data- things that we will learn more about in future- it would be easy to conclude that the datasets are roughly identical.

<img src="https://drive.google.com/uc?export=view&id=1sAVEAW5MghvfT6eAVbJTCHBKRjXLU0kB" />

### Mapping

_Mapping_ refers to the process of connecting units and variables to marks and channels. Referring back to the plot of Anscombe's quartet from above, we can recreate the mapping which generated this graphic.


The visualization maps three types of units to three types of marks. Rows in the dataset are mapped to points. Each row has two variables which we wish to visualize: x and y, which are mapped to the channels of horizontal and vertical position, respectively. The computed linear regressions are mapped to lines, with the variables of the coefficients mapped to the channels of position and angle. And finally, the full datasets are mapped to canvases, so that the first dataset appears on a different canvas from the second dataset and so on.

**Human perception is one of the strongest tools that we have for pattern recognition.**

## What is Data Visualization?

A data visualization is a _map_ from _units_ and _variables_ to _marks_ and _channels_.

*Note:* we are only going to talk about visualizing tabular data (i.e. a spreadsheet, a table, etc.). There are other kinds of data that can be visualized: networks, fields, hypercubes, and all kinds of other things. The general principles that we discuss here apply to those as well, but we may need to adjust our definition to be more general.

## Line Charts

The basic structure of a Matplotlib plot is 

    plt.figure()
    plt.plot(____________)
    plt.show()
    

- `plt.figure` sets up a figure object
- `plt.plot` determines the type of plot we are creating and takes in the data needed
- `plt.show` shows the figure on screen

The default of using `plt.plot` is to create a line plot.

## Bar Charts
The general syntax for a bar chart is:

    plt.figure()
    plt.bar( bar_names, bar_heights )
    plt.show()

We can make a simple bar chart using some sample data:

```python
plt.figure()
plt.bar(['a','b','c'], [5, 10, 3])
plt.show()
```

Let's make some bar plots using our data. Let's say we want to plot how frequently each Industry category appears.
We can use the `value_counts()` method in Pandas to get a frequency count for each. 

```python
# using value_counts() on our industry column
print(customer_df['INDUSTRY'].value_counts() )
```

Since `value.counts()` outputs a series object, we can get the industry labels by using `.index` and we can get the frequency counts by using `.values`. 

```python
print('labels: ', customer_df['INDUSTRY'].value_counts().index )
```

```python
print('values: ', customer_df['INDUSTRY'].value_counts().values )
```

We can use these 2 items to create our frequency bar chart:

```python
plt.figure()
plt.bar(customer_df['INDUSTRY'].value_counts().index, customer_df['INDUSTRY'].value_counts().values)
plt.show()
```

You can see that we have some overlap in the labels. We can rotate our labels using the `.xticks` method which controls the ticks on the x-axis. 

```python
plt.figure()
plt.bar(customer_df['INDUSTRY'].value_counts().index, customer_df['INDUSTRY'].value_counts().values)
plt.xticks(rotation=45)
plt.show()
```

We can also achieve the same type of categorical frequency plot directly from the dataframe:

```python
customer_df['INDUSTRY'].value_counts().plot(kind='bar')
plt.ylabel('Frequency in column')
plt.title('Industry count')
plt.show()
```

# Charts in Seaborn

Some types of mappings are so common and useful that they have names. Many of these are implemented easily in a package called Seaborn, so that you can quickly produce useful visualizations.

## Scatterplots

A scatterplot is a mapping of the data onto points, with one (usually continuous) variable mapped to the horizontal position, and another (usually continuous) variable mapped to the vertical position.

A useful convention is that in cases where there is a notion of an independent variable and a dependent variable, independent variables are mapped to the _horizontal position_ channel while dependent variables are mapped to the _vertical position_ channel.

```python
import seaborn as sns

sns.set_style('whitegrid')
```

```python
plt.figure(figsize = (15,10))
sns.scatterplot(data=diamonds, x='carat', y='price')
```

What does this plot tell you about the relationship between price and size?

There are lots of ways to add more information into a scatterplot. The main way is by adding another mapping from another variable to another channel. Let's add a mapping from the `clarity` variable to `hue`; seaborn makes this very easy:

```python
plt.figure(figsize = (15,10))
sns.scatterplot(data=diamonds, x='carat', y='price', hue='clarity', alpha=0.8)  
plt.show()
# lowering alpha makes points slightly more transparent
```

We can just keep adding mappings onto the scatterplot, but at a certain point it can become hard to tell what's going on. Do we learn anything from the following chart?

```python
plt.figure(figsize = (15,10))
sns.scatterplot(data=diamonds, x='carat', y='price', hue='clarity', size = 'table', style = 'color',
               sizes = (50, 100), alpha=0.8)  # lowering alpha makes points block each other a little less
```

## Bar Plots
A bar plot maps units onto rectangles. A discrete variable is mapped onto the horizontal position of the base of the rectangle, and a continuous variable is mapped onto the height of the rectangle. In order to use a bar plot, it must be the case that for each value of the discrete variable, there is only one value of the continuous variable. This suggests that you'll usually need to do some kind of split-apply-combine operation to create a bar plot.

```python
counts = diamonds.groupby('cut')['price'].agg(len)
counts.name = 'count'
```

```python
sns.barplot(y=counts, x=counts.index, order=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'])
```

A bar plot of counts is common enough that it's implemented as its own function, `catplot`. So, use that instead of doing what we wrote above.

```python
sns.catplot(data=diamonds, x='cut', kind = 'count', order=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'])
```

Another use for bar plots is to show whether different groups have different averages. Typically estimates of averages are delivered with confidence intervals alongside, and seaborn does this automatically for us. These are created by specifying `kind='bar'` in the `kind` argument.

The following block of code looks at the average price of diamonds of each cut.

```python
sns.catplot(data=diamonds, x='cut', y='price', kind = 'bar', order=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'])
```

This seems surprising! Why would the Ideal cut have the lowest average price?

```python
sns.catplot(data=diamonds[diamonds.carat > 2], x='cut', y='price', kind = 'bar', order=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'])
```

## Histograms

A histogram is a way of visualizing the entire distribution of a continuous variable in your dataset. Histograms are created by forming some number of equal width buckets such that each observation fits into exactly one bucket. A bar chart is then formed based of of counts of the buckets, producing the distribution visualization.

Histograms are implemented in seaborn with the `distplot` function. To make a histogram, pass the variable to `distplot` and set `kde=False` .

```python
sns.distplot(diamonds['price'], kde = False)
```

Omitting `kde=True` creates a slightly different type of chart called a **density plot**. A density plot is scaled so that the total area of the histogram is equal to 1, which is a way to estimate the [probability density function](https://en.wikipedia.org/wiki/Probability_density_function). Seaborn will also plot an estimate of the probability density function using something called Kernel Density Estimation (that's what `kde` stands for).

```python
sns.distplot(diamonds['price'])
```

## Line Plots

A line plot is a scatterplot with an additional mapping: _consecutiveness_ is mapped to _connectedness_. That is to say, when units appear _consecutively_ in the dataset, their corresponding marks will be _connected_. The most typical way to connect consecutive points is with a line segment.

## Facet Grids

Facet grids are another method for creating a grid or matrix of subplots. In a facet grid, the data is split up by some categorical variable and plots are created for each value of the categorical variable. 

```python
# creates a facet grid with the values of the variable "cut" mapped to rows
facet_grid = sns.FacetGrid(diamonds, row = 'cut')
```

The facet grid starts out as blank, since we haven't told it what kind of plots we want. Adding plots to the facet grid is done with the `.map()` method. Let's look at histograms of price for each cut.

```python
facet_grid = sns.FacetGrid(diamonds, row = 'cut')
facet_grid.map(sns.distplot, "price")
```

Facetting can be done across rows _and_ columns. The result will provide a lot of information.

```python
facet_grid = sns.FacetGrid(diamonds, row = 'cut', col = 'clarity')
facet_grid.map(sns.distplot, "price")
```

### Graphing Functions

The examples we have looked at involve us graphing discrete data sets.  How would we try to graph a continuous function, like $\sin(x)$, or $e^{-x^2/2}$?  The answer is we have to create a series of data points for Matplotlib to use as the x-axis, fine enough so that when it interpolates between the points with lines, it's not noticeable to the eye.

To accomplish this, we're going to use the `np.arange()` function (note:  it's not "arrange", it's a range).  This function will produce a sequence of evenly space points, with syntax `np.arange(start,end,step)`.

```python
np.arange(0,10,0.5)
```

```python
t = np.arange(0.0, 1.0+0.01, 0.01)
s = np.sin(2*np.pi * t)
sns.lineplot(x=t, y=s)
```

What would happen if we had changed the step size from 0.01 to 0.1?

We could also make it look more like a graph appearing in a math textbook in the same way we've adjusted other graphs:

```python
t = np.arange(0.0, 1.0+0.01, 0.01)
s = np.sin(2*np.pi * t)
sns.lineplot(x=t, y=s)
```

Keeping the step to a hundredth of the total range works well, as a rule of thumb, to keep the graph looking smooth:

```python
t = np.arange(1.0, 100.0+1, 1)
l = np.log(t/2)
sns.lineplot(x=t, y=l)
```

Compare to the step size a tenth of the one above:

```python
t = np.arange(1.0, 100.0+0.1, 0.1)
l = np.log(t/2)
sns.lineplot(x=t, y=l)
```

These tools will allow us to graph both data sets and mathematical functions on the same graph, which will be a good way to visually inspect any models we create.

## Bokeh

Most of the plotting we will do in this course will be with Matplotlib, but we will have a short look at Bokeh with the aim of easily creating interactive plots.  This graphing package comes with many tools for interactivity, including zooming, selecting, panning, and more.  This will let us get a "closer look" at the data we're graphing.

```python
from bokeh.plotting import figure, show, output_file
from bokeh.io import output_notebook
from bokeh.models import DatetimeTickFormatter
```

```python
output_notebook()
```

The general syntax for a Bokeh plot is as follows:
1. Define the `TOOLS` you want (e.g. pan, wheel_zoom, etc.).
2. Declare a variable name for the `figure()`, like we did with `plt.figure()`.
3. Feed `figure()` arguments for your graph, such as `title`, `plot_width` or `height`, and `tools=TOOLS`.
4. Define the `xaxis` and `yaxis` `.axis_labels`.
5. Decide what kind of plot you want (e.g.: `line`,`scatter`, etc.), and declare what your x and y axes will be.
6. Call `show(variable_name)` to display the graph

So if we were going to plot the open prices from `df2` as a line graph, we would write:

```python
import pandas as pd

df2 = pd.read_csv('goog.csv')

TOOLS = "pan,wheel_zoom,box_zoom,reset,save"

p1 = figure(title="Futures Prices", tools=TOOLS,x_axis_type='datetime', plot_width=900, plot_height=700)
p1.xaxis.axis_label = 'Date'
p1.yaxis.axis_label = 'Price'
p1.line(df2.index, df2['Open'], legend_label="Open")
show(p1)
```

Try moving the graph around, zooming with box-zoom or wheel zoom, and then using the reset button to return the graph to its original position.

We can graph high and low again using:

```python
TOOLS = "pan,wheel_zoom,box_zoom,reset,save"

p1 = figure(title="Futures Prices", tools=TOOLS,x_axis_type='datetime', plot_width=900, plot_height=700)
p1.xaxis.axis_label = 'Date'
p1.yaxis.axis_label = 'Price'
p1.line(df2.index, df2['High'], legend_label="High",color='red')
p1.line(df2.index, df2['Low'], legend_label="Low",color='blue')
show(p1)
```

We can also put plots side-by-side again, using the `gridplot` from `bokeh.layouts`.  We define all of our figures, and then when we call the `show()` function, we pass it a list of lists of our plots.  

```python
from bokeh.layouts import gridplot

TOOLS = "pan,wheel_zoom,box_zoom,reset,save"

p1 = figure(title="Futures Prices", tools=TOOLS,x_axis_type='datetime')
p1.xaxis.axis_label = 'Date'
p1.yaxis.axis_label = 'Price'
p1.line(df2.index, df2['High'], legend_label="High")

p2 = figure(title="Futures Prices", tools=TOOLS,x_axis_type='datetime')
p2.xaxis.axis_label = 'Date'
p2.yaxis.axis_label = 'Price'
p2.line(df2.index, df2['Low'], legend_label="Low")


show(gridplot([[p1,p2]], plot_width=400, plot_height=300))
```

Again, this sort of plotting barely scratches the surface of what Bokeh is capable of.  Here is one of the front-page examples from their gallery, creating a scatter plot of different sized discs with a range of colors:

```python
N = 4000
x = np.random.random(size=N) * 100
y = np.random.random(size=N) * 100
radii = np.random.random(size=N) * 1.5
colors = [
    "#%02x%02x%02x" % (int(r), int(g), 150) for r, g in zip(50+2*x, 30+2*y)
]

TOOLS="hover,crosshair,pan,wheel_zoom,zoom_in,zoom_out,box_zoom,undo,redo,reset,tap,save,box_select,poly_select,lasso_select,"

p = figure(tools=TOOLS,plot_width=800, plot_height=800)

p.scatter(x, y, radius=radii,
          fill_color=colors, fill_alpha=0.6,
          line_color=None)
show(p)
```
